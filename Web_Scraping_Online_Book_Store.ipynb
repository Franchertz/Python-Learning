```python
#number of pages to be selected
no_of_pages=50

#looping through the required pages and selecting the pages accordingly
for i in range(1,no_of_pages+1):
  url=('http://books.toscrape.com/catalogue/page-{}.html'.format(i))    #to include the last page
```

```python
pages.append(url)  #adding all the content of the respective pages to the created page list
```

```python
print('Number of pages:', len(pages))
print(pages)
```

**Getting the Data from the particular page url and converting from object of type 'requests' to object of type 'beautifulsoup'**

making it more readable

```python
import requests

import bs4

import pandas as pd
```

```python
for item in pages:
    page=requests.get(item)
    soup=bs4.BeautifulSoup(page.text,'html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

#*Finding all the Titles*

```python
#print all the titles

for t in soup.findAll('h3'):
  print(t)
```

```python
#After finding all the titles, we now want to add them to the empty list created earlier

for t in soup.findAll('h3'):
   titless=t.getText()
   title.append(titless)
```

#*Finding all the Price*

```python
#Finding all the prices

for p in soup.find_all('p', class_='price_color'):  #the 'p' tag is in the 'price color' class
    price=p.getText()
    prices.append(price)
```

#*Let's get the Ratings*

```python
#Getting the atings

for s in soup.find_all('p',class_='star-rating'):
  for k,v in s.attrs.items():
    star=v[1]
    ratings.append(star)
    print(star)
```

#*Finding all the URL's of the images*

```python
#To fetch all the div tags in the class called image_container

divs=soup.find_all('div',class_='image_container')
#print(divs)
for thumbs in divs:
  tagss=thumbs.find('img',class_='thumbnail')
  #print(tagss)
  links='http://books.toscrape.com/' + str(tagss['src'])
  newlinks=links.replace('..','')
  urls.append(newlinks)
```
