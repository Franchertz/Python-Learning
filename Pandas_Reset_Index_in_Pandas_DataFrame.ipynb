**Rename a Column In Pandas DataFrame**

```python
my_dataframe.rename(columns={'column1':'first', 
                             'column2':'second', 
                             'column3':'third', 
                             'column5':'fourth',
                             'column6':'fifth'}, 
                    inplace=True)  # using the function rename() the columns
```

```python
my_dataframe
```

```python
my_dataframe.rename(columns={'first':1}) # renaming a single column with the function rename()
```

**Rename Index In Pandas DataFrame**
