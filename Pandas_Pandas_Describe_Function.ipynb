**Let's find which car is having price greater than 20,000**

```python
data_auto[data_auto['price']>20000][['make','price']] # This function gives which car cost greater than 20000
```

**Let's see all Volvo cars that cost less than 50,000**

```python
data_auto[(data_auto['make']=='volvo') & (data_auto['price']<50000)]
```

**For Make and Price columns, see below**

```python
data_auto[(data_auto['make']=='volvo') & (data_auto['price']<50000)][['make','price']]
```

```python
data_auto[(data_auto['body_style']=='sedan') & (data_auto['price']<50000)][['body_style','price']]
```

```python
data_auto[(data_auto['body_style']=='convertible') & (data_auto['price']<50000)][['body_style','price']]
```
