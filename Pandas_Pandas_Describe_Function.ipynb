#**Join & Merge:Creating Dataset**

```python
# DataFrame of number of sales made by employees
sales = {'Jones': 10000,
         'Chris': 5000,
         'Piyush': 440,
         'Meera': 6700,
         'Rahul': 300
         }

# DataFrame of all employees and the region they work in
region = {'Jones': 'West',
          'Chris': np.nan,
          'Piyush': 'West',
          'Meera': np.nan,
          'Anthony': 'East',
          'Ellen': 'South',
          'Josh': 'West',
          'Simran': 'East',
          'Oscar': 'North'
          }
```

```python
#Converting dictionary to DataFrames
sales_df = pd.DataFrame.from_dict(sales, orient='index',
                                  columns=['sales'])

region_df = pd.DataFrame.from_dict(region, orient='index',
                                   columns=['region'])
```

```python
sales_df
```

```python
region_df
```

#**Join**

*Inner Join*

*Left Outer Join*

*Right Outer Join*

*Full Outer Join*

```python
#    it treats the region_df as left table and the sales_df as right table
#    and therefore all the items in region_df will appear and only
#    those items in sales_df that match with region_df will appear

joined_df = region_df.join(sales_df, how='left')
print(joined_df)
```

```python
#    it treats the region_df as left table and the sales_df as right table
#    and therefore all the items in sales_df will appear and only
#    those items in region_df that match with region_df will appear

joined_df = region_df.join(sales_df, how='right')
print(joined_df)
```

```python
joined_df = region_df.join(sales_df, how='inner') #Showing Inner Join
print(joined_df)
```

```python
#  Full outer join, ie joining all the data points in both DataFrames
joined_df = region_df.join(sales_df, how='outer') 
print(joined_df)
```

#**Merge**

```python
# give title to the index column
# so the dataframes can be merged on a commonality ie names

region_df.index.name='names'
sales_df.index.name='names'
```

```python
print(region_df)

print('-'*10)

print(sales_df)
```

```python
merged_df = pd.merge(region_df,sales_df, on='names') #merge on a key
```

```python
print(merged_df)
```

**Or a simple code that prints out everything**

```python
print(pd.merge(region_df,sales_df, on='names'))
```

```python
print(pd.merge(region_df,sales_df, on='names', how='left')) #Same as doing a Left join
```

```python
print(pd.merge(region_df,sales_df, on='names', how='right')) #Same as doing a right join
```

```python
print(pd.merge(region_df,sales_df, on='names', how='outer')) #Same as doing a full outer join
```
